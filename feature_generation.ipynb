```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import re
```

## Incorprate AOA

```python
aoa_df = pd.read_csv('AOA_51715_words.csv', header = 0, index_col = 0)
aoa_df.head()
```

```python
aoa_df.info()
```

```python
train_data = pd.read_csv('WikiLarge_Train.csv', header = 0)
train_data.head()
```

```python
def text_processing(x):
    x = re.sub(r'[^\w\s]','', x)
    x = x.split(' ')
    x = [word.strip().lower() for word in x]
    return x
```

```python
train_data.loc[:, 'processed_text'] = train_data['original_text'].apply(text_processing)
```

```python
train_data.head()
```

```python
def calculate_average_aoa(x):
    aoa_value_list = []
    
    for word in x:
        if word:
            try:
                aoa_value = aoa_df.loc[word, 'AoA_Kup_lem']
                if aoa_value:
                    aoa_value_list.append(aoa_value)
            except:
                pass
    
    if aoa_value_list:
        mean_aoa = np.mean(aoa_value_list)
    else:
        mean_aoa = np.nan
        
    return mean_aoa
```

```python
train_data.loc[:, 'average_AoA_Kup_lem'] = train_data['processed_text'].apply(calculate_average_aoa)
```

```python
train_data.head()
```

```python
train_data.info()
```

## Incorprate percentage of basic words

```python
with open('dale_chall.txt') as file_obj:
    basic_word_list = file_obj.readlines()
    cleaned_basic_word_list = [word.lower().strip() for word in basic_word_list]
```

```python
len(cleaned_basic_word_list)
```

```python
def calculate_percentage_of_basic_words(x):
    count = 0
    for word in x:
        if word and word in cleaned_basic_word_list:
            count += 1
    
    percentage = count / len(x)
    
    return percentage
```

```python
train_data.loc[:, 'percentage_of_basic_words'] = train_data['processed_text'].apply(calculate_percentage_of_basic_words)
```

```python
train_data.head()
```

## Incorprate Concreteness

```python
concreteness_df = pd.read_csv('Concreteness_ratings_Brysbaert_et_al_BRM.txt', sep = '\t', header = 0, index_col = 0)
concreteness_df
```

```python
concreteness_df.info()
```

```python
def calculate_average_concrete_mean(x):
    aoa_value_list = []
    
    for word in x:
        if word:
            try:
                aoa_value = concreteness_df.loc[word, 'Conc.M']
                if aoa_value:
                    aoa_value_list.append(aoa_value)
            except:
                pass
    
    if aoa_value_list:
        mean_aoa = np.mean(aoa_value_list)
    else:
        mean_aoa = np.nan
        
    return mean_aoa
```

```python
train_data.loc[:, 'average_concreteness'] = train_data['processed_text'].apply(calculate_average_concrete_mean)
```

```python
train_data.head()
```

```python
def calculate_percentage_known_mean(x):
    aoa_value_list = []
    
    for word in x:
        if word:
            try:
                aoa_value = concreteness_df.loc[word, 'Percent_known']
                if aoa_value:
                    aoa_value_list.append(aoa_value)
            except:
                pass
    
    if aoa_value_list:
        mean_aoa = np.mean(aoa_value_list)
    else:
        mean_aoa = np.nan
        
    return mean_aoa
```

```python
train_data.loc[:, 'average_percentage_known'] = train_data['processed_text'].apply(calculate_percentage_known_mean)
```

```python
train_data.head()
```

```python
train_data.to_csv('train_data_with_feature.csv', index = False)
```

```python
def generate_features(train_data):
    train_data.loc[:, 'processed_text'] = train_data['original_text'].apply(text_processing)
    train_data.loc[:, 'average_AoA_Kup_lem'] = train_data['processed_text'].apply(calculate_average_aoa)
    train_data.loc[:, 'percentage_of_basic_words'] = train_data['processed_text'].apply(calculate_percentage_of_basic_words)
    train_data.loc[:, 'average_concreteness'] = train_data['processed_text'].apply(calculate_average_concrete_mean)
    train_data.loc[:, 'average_percentage_known'] = train_data['processed_text'].apply(calculate_percentage_known_mean)
```
